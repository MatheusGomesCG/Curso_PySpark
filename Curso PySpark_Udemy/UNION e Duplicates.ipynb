```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
spark = (
    SparkSession.builder.master('local[*]')
    .appName('Aula de Union e Drop')
    .config('spark.sql.legacy.timeParserPolicy', 'LEGACY')
    .getOrCreate()
)
```

```python
spark
```

```python
df = spark.read.parquet('./DATASETS/LIVROS.parquet').limit(5)
```

```python
df.show()
```

```python
# Tem o intuito de unir dois dataframes (coloca um sobre o outro)
dup = df.union(df)
dup.show()
```

```python
df2 = dup.unionByName(df, allowMissingColumns=True)
df2.show()
```

```python
dup.drop_duplicates().show()
```

```python
df4 = spark.read.parquet('./DATASETS/LIVROS.parquet').where(F.col('numero_paginas').isin(461, 209))
df4.show()
```

```python
df4.drop_duplicates(['numero_paginas']).show()
```

```python
# Relembrando os conceitos de SQL
df4.groupBy(F.col('numero_paginas')).agg(F.count('*').alias('qnt')).orderBy(F.asc(F.col('numero_paginas'))).show()
```
